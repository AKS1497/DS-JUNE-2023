### 3. Research Question on Hypothesis Testing

In previous years, 52% of parents believed that electronics and social
media was the cause of their teenager’s lack of sleep. Do more parents
today believe that their teenager’s lack of sleep is caused due to
electronics and social media?

**Population**: Parents with a teenager (age 13-18)  
**Parameter of Interest**: p

**Null Hypothesis:** p = 0.52  
**Alternative Hypthosis:** p \> 0.52 (note that this is a one-sided
test)

**Data**: 1018 people were surveyed. 56% of those who were surveyed
believe that their teenager’s lack of sleep is caused due to electronics
and social media.

### Hint: Use `proportions_ztest()` from `statsmodels`

Note the argument `alternative="larger"` indicating a one-sided test.
The function returns two values - the z-statistic and the corresponding
p-value.

### What is your Conclusion of the hypothesis test

import statsmodels.api as sm

Given data

sample_size = 1018

sample_proportion = 0.56 \# 56% believe in electronics and social media

Null hypothesis

null_proportion = 0.52 \# 52% believed in the previous years

z_statistic, p_value = sm.stats.proportions_ztest(sample_proportion \*
sample_size, sample_size, null_proportion, alternative='larger')

p-value

alpha = 0.05 \# Set your significance level

if p_value \< alpha:

conclusion = "Reject the null hypothesis"

else:

conclusion = "Fail to reject the null hypothesis"

print(f"Z-Statistic: {z_statistic}")

print(f"P-Value: {p_value}")

print("Conclusion:", conclusion)

### 5. Calculate a function to generate random arrays with range of (1,100) and the naive functions to calculate Mean, Varience and Standard deviation for the array generated


    import random
    import math

     
    def generate_random_array(length, min_value, max_value):
        return [random.randint(min_value, max_value) for _ in range(length)]


    def calculate_mean(arr):
        return sum(arr) / len(arr)


    def calculate_variance(arr):
        mean = calculate_mean(arr)
        variance = sum((x - mean) ** 2 for x in arr) / len(arr)
        return variance


    def calculate_std_deviation(arr):
        variance = calculate_variance(arr)
        std_deviation = math.sqrt(variance)
        return std_deviation


    random_array = generate_random_array(20, 1, 100)
    print("Random Array:", random_array)

    mean = calculate_mean(random_array)
    variance = calculate_variance(random_array)
    std_deviation = calculate_std_deviation(random_array)

    print("Mean:", mean)
    print("Variance:", variance)
    print("Standard Deviation:", std_deviation)
