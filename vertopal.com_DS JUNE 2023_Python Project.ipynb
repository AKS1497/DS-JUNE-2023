    it yields a 3-tuple (dirpath, dirnames, filenames)

dirpath: is a string, the path to the directory dirnames: is a list of
the names of the subdirectories in dirpath filenames: is a list of the
names of the non-directory files in dirpath The fnmatch module compares
file names against glob-style patterns.

fnmatch() compares a single file name against a pattern and returns a
boolean indicating whether or not they match.
